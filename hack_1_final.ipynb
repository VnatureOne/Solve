```python
from pyqubo import Binary, Spin
```

```python
stocks = pd.read_csv("task1.csv")
```

```python
stocks
```

```python
N = 10
M = 1000  - 2 ** (N - 1)
```

```python
M
```

```python
cov_matrix = stocks.cov().to_numpy()
```

```python
X = np.zeros((100, N), dtype=object)
```

```python
for i in range(100):
  for j in range(N):
    X[i, j] = Binary(f"{i}, {j}")
X
```

```python
w = np.zeros((100, ), dtype=object)
```

```python
for i in range(100):
  w[i] = sum(0.001 * (2 ** q) * X[i, q] for q in range(N - 1)) + 0.001 * M * X[i, -1]
```

```python
log_returns_arr = []
for i in stocks:
 log_returns_arr.append((np.log(stocks[i] / stocks[i].shift(1))).to_list())
```

```python
log_returns_arr = np.array(log_returns_arr)
nu = np.zeros((100, ))
for i in range(100):
  nu[i] = np.sum(log_returns_arr[i, 1:]) / 99
```

```python
ro = Placeholder("ro")
gamma = Placeholder("gamma")
alpha = Placeholder("alpha")
C = ro * (sum(w) - 1) ** 2
B = gamma * (np.dot(w, np.dot(cov_matrix, w)))
A = - alpha * np.dot(nu, w)
H = A + B + C
```

```python
model = H.compile()
feed_dict = {"gamma": 7, "ro": 1, "alpha": 1}
qubo, qubo_offset = model.to_qubo(feed_dict=feed_dict)
```

```python
from dwave.system import EmbeddingComposite, DWaveSampler
from neal import SimulatedAnnealingSampler

# 1. Сформируем объект сэмплера
sampler = SimulatedAnnealingSampler()

# 2. Выполним отжиг
sampleset = sampler.sample_qubo(qubo, num_reads=100)

# 3. Получим минимальное значение и соответствующую конфигурацию
best_sample = sampleset.first.sample
best_energy = sampleset.first.energy

print("Лучший результат:", best_sample)
```

```python
Q = np.zeros(())
```

```python
res = []
for n in range(100):
  s = 0
  for i in best_sample:
    if int(i.split(",")[0]) == n:
      if int(i.split(",")[1]) == 9:
        s += best_sample[i] * M
      else:
        s += (2 ** int(i.split(",")[1]) * best_sample[i])
  res.append(s)
```

```python
import math
```

```python
N = []
for i in range(len(res)):
  N.append(math.floor(res[i] * 1000 / stocks.loc[0].values[i]))
```

```python
start_price = np.dot(N, stocks.loc[0].values)
start_price
```

```python
R = 0
for i in range(99):
  P_current = np.dot(N, stocks.loc[i].values)
  P_next = np.dot(N, stocks.loc[i + 1].values)
  r = (P_next - P_current) / P_current
  R += r
R /= 99
R
```

```python
sum = 0
for i in range(99):
  P_current = np.dot(N, stocks.loc[i].values)
  P_next = np.dot(N, stocks.loc[i + 1].values)
  r = (P_next - P_current) / P_current
  sum += (r - R) ** 2
sigma = np.sqrt(99 / 98 * sum)
sigma
```

```python
start_money = np.dot(N, stocks.loc[0].values)
end_money = np.dot(N, stocks.loc[99].values)
final_profit = (end_money - start_money) / start_money
final_profit
```
